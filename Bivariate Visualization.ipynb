# Bivariate Visualization

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('fuel-econ.csv')
```

```python
df.head()
```

### Scatterplot : Quantitative & Quantitative Variables
### Violinplot: Quantitative & Categorical Variables
### Clustered Bar Chart: Categorical & Categorical Variables

```python
fig, ax = plt.subplots()
plt.yticks([5,10,15,20,25,30])
plt.xticks([1,2,3,4,5,6])
plt.xlabel('Hot Coffee')
plt.ylabel('Temperature')
plt.show()
```

```python
df.head()
```

```python
plt.scatter(data=df, x='displ',y='comb')
plt.xlabel('Displacement(L)')
plt.ylabel('Comb Fuel Eff (mpg)')
plt.title('Bivariate Plot');
```

```python
sns.regplot(data=df, x='displ',y='comb')
plt.xlabel('Displacement(L)')
plt.ylabel('Comb Fuel Eff (mpg)');
```

## Overplotting, Transparency, Jitter

```python
sns.regplot(data=df, x='year',y='comb', x_jitter=0.3, scatter_kws={'alpha':0.3})
plt.xlabel('Year')
plt.ylabel('Comb Fuel Eff (mpg)');
```

### Heat Map

```python
plt.hist2d(data=df, x='displ',y='comb', cmin=0.5, cmap='viridis_r')
plt.colorbar();
```

### Violin Plot & Box Plot

```python
plt.figure(figsize=[20,5])
plt.subplot(1,2,1)
sns.violinplot(data=df, x='VClass',y='comb')
plt.xticks(rotation=15)

plt.subplot(1,2,2)
sns.boxplot(data=df, x='VClass',y='comb')
plt.xticks(rotation=15);
```

### Clustered Bar Chart

```python
sns.countplot(data=df, x='VClass');
```

```python
df.head()
```

```python
df['trans_type'] = df['trans'].apply(lambda x:x.split()[0])
```

```python
df.head()
```

```python
sns.countplot(data=df, x='VClass', hue='trans_type')
plt.xticks(rotation=15);
```

### Faceting

```python
a = sns.FacetGrid(data=df, col='VClass')
a.map(plt.hist, 'comb');
```
